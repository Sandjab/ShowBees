```python
#exemple of contextlib usage
import sqlite3
import contextlib

# Nice
def query1(statement):
    with contextlib.closing(sqlite3.connect(path_to_file)) as conn: # auto-closes
            # auto-closes ( needed because cursor is not a context manager)
            with contextlib.closing(conn.cursor()) as cursor: 
                cursor.execute(statement)

# Bad
def query2(statement):
    with sqlite3.connect(path_to_file)) as conn: # auto commit only
        with conn: # useless, same as above
            with conn.cursor() as cursor: # does nothing
                cursor.execute(statement)

# bad
def query3(statement):
    with sqlite3.connect(path_to_file)) as conn: # auto commit only
        with conn.cursor() as cursor: # does nothing
            cursor.execute(statement)
        conn.close()    

# Ok
def query5 (statement):
    con = sqlite3.connect(path_to_file)
    with con:
        cur = con.cursor()
        cur.execute(statement)
        res = cur.fetchall()
    if con:
        con.close()

    return res                
```
