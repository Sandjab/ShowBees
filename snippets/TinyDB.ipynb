```python
from tinydb import TinyDB, Query
with TinyDB('db.json') as tdb:
    tdb.truncate()
    tdb.insert({'type': 'apple', 'count': 7})
    tdb.insert({'type': 'peach', 'count': 3})
    tdb.insert({'type': 'apple', 'count': 7})
    tdb.insert({'type': 'orange', 'count': 7})
    tdb.insert({'type': 'peach', 'count': 2, 'ext': 'bof'}) 
    #print(tdb.all())
    
    for item in tdb:
        print(item['type'])
        
    print('-------------------------------------------------------------')    
        
    q = Query()
    result = tdb.search(q.type == 'peach' and q.ext == 'bof')
    for item in result:
        print(item)
        
        
    tdb.update({'count': 69}, q.type == 'orange')    
    
    for item in tdb:
        print(item)    
```
