```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np 

sig, sr = librosa.core.load('samples/violin.wav', sr = 22050)


mfccs = librosa.feature.mfcc(y=sig ,sr=sr, n_mfcc = 40)
print(mfccs.shape[0])
print(mfccs.shape[1])
#print(x)


librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
```
