```python
import time
import librosa

SAMPLE_RATE = 22050
DURATION = 1
#SOUNDFILE = "samples/test600s.mp3"
SOUNDFILE = "samples/test595s.wav"


start_time0 = time.time()
source, sr = librosa.core.load(SOUNDFILE, sr=SAMPLE_RATE);
end_time = time.time()
print("LIBROSA FULL MP3:", end_time - start_time0)
print("L =", len(source))

for i in range( int(len(source)/(sr*DURATION)) ):
    start_time = time.time()
    librosa.output.write_wav("tmp/chunk" + str(i).zfill(3) + ".wav", source[i* sr * DURATION:(i + 1) * sr * DURATION], sr)
    end_time = time.time()
    print("WRITEWAV",i,":", end_time - start_time)
end_time = time.time()
print("--------------------------")
print("TOTAL DURATION:", end_time - start_time0)


```
