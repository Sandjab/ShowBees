```python
# Display cardinalities by hives and queen/noqueen
sql = """
    select distinct a.hive, l.queen, count(*)
    from attributes a, labels l
    where a.sample_id = l.sample_id
    group by a.hive, l.queen
    order by a.hive
    """
ds.requestDataFrame(sql)
```

```python
sql = """
select s.name, s.file_id, a.hive, l.queen from samples s, attributes a, labels l
where a.sample_id = s.rowid
and l.sample_id = s.rowid
and l.nobee = 0
"""
df = ds.requestDataFrame(sql)
print(df)
```
